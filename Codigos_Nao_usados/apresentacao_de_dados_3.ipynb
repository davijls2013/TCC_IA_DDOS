```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
# Caminho para o arquivo CSV no seu SSD
csv_file_path = 'DrDoS_UDP.csv'

# Definir a quantidade de linhas para pular durante a leitura (ajuste conforme necessário)
skip_rows = 100

# Função para plotar diferentes tipos de gráficos
def plot_graph(graph_type, x_data, y_data, x_label, y_label, title):
    plt.figure()
    if graph_type == 'line':
        plt.plot(x_data, y_data)
    elif graph_type == 'scatter':
        plt.scatter(x_data, y_data)
    # Adicione mais opções para outros tipos de gráficos

    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.title(title)
    plt.show()

# Leitura do arquivo CSV em partes (chunks) para otimização de uso de memória
chunk_size = 10000
data_frames = []
```

```python
for chunk in pd.read_csv(csv_file_path, chunksize=chunk_size):
    data_frames.append(chunk)

pd.set_option('display.max_columns', None)

# Concatenar os dataframes lidos em um único dataframe
full_data_frame = pd.concat(data_frames, ignore_index=False)
```

```python
metric_column_mapping = {
    'Gráfico de Linhas Temporais Combinado': ['Flow Packets/s', 'Flow Packets/s', 'Packet Length Mean'],
    'Gráfico de Dispersão 2D': ['Flow Packets/s', 'Packet Length Mean'],
    'Gráfico de Barras Agrupadas': ['Flow Packets/s', 'Flow Packets/s', 'Packet Length Mean'],
    'Gráfico de Área Empilhada': ['Flow Packets/s', 'Flow Packets/s', 'Fwd PSH Flags', 'Bwd PSH Flags'],
    'Gráfico de Bolhas': ['Flow Packets/s', 'Packet Length Mean', 'Flow IAT Std'],
    'Gráfico de Radar': ['Flow Packets/s', 'Flow Packets/s', 'Packet Length Mean'],
    'Gráfico de Mapa de Calor': ['Flow Packets/s', 'Flow Packets/s', 'Packet Length Mean']
    # Adicione mais correspondências para outros gráficos
}
```

```python
# Gerar gráficos para cada métrica e colunas correspondentes
for metric, columns in metric_column_mapping.items():
    graph_type = metric.split()[2]  # Obtém o tipo de gráfico do nome da métrica
    x_col, y_col = columns[:2]
    print(full_data_frame)
    print(x_col)
    x_data = full_data_frame[str(x_col.strip())]
    y_data = full_data_frame[str(y_col.strip().str())]
    title = f'{metric}:\n{x_col} vs. {y_col}'
    plot_graph(graph_type, x_data, y_data, x_col, y_col, title)
```
