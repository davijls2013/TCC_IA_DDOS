```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from itertools import combinations
```

```python
# Caminho para o arquivo CSV no seu SSD
csv_file_path = 'DrDoS_UDP.csv'

# Definir a quantidade de linhas para pular durante a leitura (ajuste conforme necessário)
skip_rows = 100
```

```python
# Função para plotar o gráfico de linha da relação entre duas colunas
def plot_relation(x_column, y_column, x_data, y_data):
    plt.plot(x_data, y_data)
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.title(f'Relação entre {x_column} e {y_column}')
    plt.show()
```

```python
# Leitura do arquivo CSV em partes (chunks) para otimização de uso de memória
chunk_size = 10000
data_frames = []

for chunk in pd.read_csv(csv_file_path, chunksize=chunk_size):
    data_frames.append(chunk)

pd.set_option('display.max_columns', None)
```

```python
# Concatenar os dataframes lidos em um único dataframe
full_data_frame = pd.concat(data_frames, ignore_index=False)

# Obter todas as combinações possíveis de colunas em pares
column_combinations = combinations(full_data_frame.columns, 2)
```

```python
print(full_data_frame.head)
```

```python
# Configurar a matriz de gráficos
num_columns = len(full_data_frame.columns)
fig, axs = plt.subplots(num_columns, num_columns, figsize=(10, 10), squeeze=False)
```

```python
# Preencher a matriz de gráficos
for i, (x_col, y_col) in enumerate(column_combinations):
    ax = axs[i // num_columns, i % num_columns]
    plot_relation(x_col, y_col, full_data_frame[x_col], full_data_frame[y_col])
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
```

```python
plt.tight_layout()
plt.show()
```
